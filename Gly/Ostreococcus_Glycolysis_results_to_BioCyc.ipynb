```python
import pandas as pd
import numpy as np
import re, os
def to_chr(  matchobj ):
        return chr(int(matchobj.group(1)))
def to_ord( matchobj ):
        return '__{}__'.format(ord(matchobj.group(0)))    
def munge( sbmlid ):
    alphaRE = re.compile(r'[^A-Za-z_]')
    m = alphaRE.match(sbmlid[0])
    if m:
        sbmlid = '_{}'.format(sbmlid)
    alphanumRE = re.compile(r'[^A-Za-z0-9_]')
    sbmlid = alphanumRE.sub(to_ord, sbmlid)
    return sbmlid
def unmunge( sbmlid ):
    ordRE = re.compile('__([0-9]+)__')
    if '_' == sbmlid[0] and '_' != sbmlid[1]:
        sbmlid = sbmlid[1:]
    return ordRE.sub( to_chr, sbmlid)    
def unmunge_model( model ):     
    for met in model.metabolites:
        met.id = unmunge(met.id)
    for rxn in model.reactions:
        rxn.id = unmunge(rxn.id)
    return model
```
